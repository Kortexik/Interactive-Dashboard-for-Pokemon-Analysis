```python
import pandas as pd
import numpy as np
```

```python
mons = pd.read_csv('..\\data\\pokemon.csv')
mons.info()
```

```python
#Legendary pokemons to non-legendary ones comparision
legends = mons[mons['is_legendary'] == 1]
normal = mons[mons['is_legendary'] == 0]
```

```python
mean_l = legends['base_total'].mean()
mean_n = normal['base_total'].mean()
max_base_stat = mons['base_total'].max()
filt = (mons['base_total'] == max_base_stat)
max_stat = mons[filt]['name']
```

```python
print(f'Legendary pokemons have a base stat total bigger on average by {mean_l - mean_n}')
```

```python
print(f'Pokemons with the maximum base stat: \n{max_stat} \nbase total = {max_base_stat}')
```

```python
#Stats by type
type_names = mons['type1'].unique()
type_names
```

```python
def findStatMean(typ:str):
    filt = (mons['type1'] == typ) | (mons['type2'] == typ)
    stat_mean = mons[filt]['base_total'].mean()
    return stat_mean
```

```python
stats_by_type = pd.DataFrame(type_names, columns=["type"])
stats_by_type['stat_mean'] = stats_by_type['type'].apply(findStatMean).round(2)
stats_by_type.set_index('type', inplace=True)
```

```python
stats_by_type.sort_values(by='stat_mean', ascending=False)
```

```python

gens = mons.groupby('generation')['base_total'].mean()
gens
mons
```

```python
against_columns = mons.loc[:, 'against_bug':'against_water']

def count_weaknesses(row):
    counts = row.value_counts()
    return pd.Series({
        'count_0': counts.get(0.0, 0),
        'count_0.25': counts.get(0.25, 0),
        'count_0.5': counts.get(0.5, 0),
        'count_1.0': counts.get(1.0, 0),
        'count_2.0': counts.get(2.0, 0),
        'count_4.0': counts.get(4.0, 0)
    })

weakness_counts = against_columns.apply(count_weaknesses, axis=1)

weakness_df = pd.concat([mons['name'], weakness_counts], axis=1)
weakness_df.columns = ['Name', 'x0', 'x0.25', 'x0.5','x1.0', 'x2.0', 'x4.0'] 
```

```python
def concat_type(x):
    if pd.isna(x['type2']):
        return x['type1']
    else:
        return f"{x['type1']}/{x['type2']}"


mons['type'] = mons.apply(concat_type, axis=1)
weakness_df['type'] = mons['type']
```
